```python
【分段函數】【if、elif、else 結構寫法】
        3x - 5  (x > 1)
f(x) =  x + 2   (-1 <= x <= 1)
        5x + 3  (x < -1)

x = float(input('x = '))
if x > 1:
    y = 3 * x - 5
elif x >= -1:
    y = x + 2
else:
    y = 5 * x + 3
print('f(%.2f) = %.2f' % (x, y))
```

```python
练习1：英制单位英寸与公制单位厘米互换
"""
英制单位英寸和公制单位厘米互换

Version: 0.1
Author: 骆昊
"""
value = float(input('请输入长度: '))
unit = input('请输入单位: ')
if unit == 'in' or unit == '英寸':
    print('%f英寸 = %f厘米' % (value, value * 2.54))
elif unit == 'cm' or unit == '厘米':
    print('%f厘米 = %f英寸' % (value, value / 2.54))
else:
    print('请输入有效的单位')
```

```python
value = float(input('請輸入長度:'))
unit = input('請輸入單位in/mm:')
if unit == 'in' or unit == '英吋':
    print('%.2f英吋 = %.2f公厘' %(value, value * 25.4))
elif unit == 'mm' or unit == '公厘':
    print('%.2f公厘 = %.2f英吋' %(value, value / 25.4))
else:
    print('請輸入有效單位')
```

```python
a = float(input('a='))
b = float(input('b='))
c = float(input('c='))
unit = input('請輸入單位(mm/cm)')

p = (a + b + c) / 2
area = (p * (p - a) * (p - b) * (p - c)) ** 0.5

if a + b > c and a + c > b and b + c > a:
    print('周長: %.2f' %(a + b + c))
    print('面積: %.2f' %(area))
else:
    print('無法構成三角形')
```
